```python
import csv

input_file = "./data/trajectories/go_hover_traj.csv"

perching = []
with open(input_file, newline='') as csvfile:
    reader = csv.reader(csvfile)
    header = next(reader)  # 跳过表头
    for row in reader:
        # 跳过空行
        if not row or all(cell.strip() == "" for cell in row):
            continue
        # 转为 float
        perching.append([float(cell) for cell in row])

# 打印为你想要的格式
print("go_hover_traj = [")
for item in perching:
    print("    [")
    print(",".join(str(x) for x in item))
    print("    ],")
print("]")
```
